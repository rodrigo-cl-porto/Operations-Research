# PROGRAMA√á√ÉO N√ÉO LINEAR

Os problemas de Programa√ß√£o N√£o-Linear (PNL) dizem respeito aos problemas de otimiza√ß√£o que atendem a pelo menos um dos requisitos abaixo:

- A fun√ß√£o objetivo n√£o √© uma fun√ß√£o linear, ou seja, n√£o √© expressa como $\text{Max ou Min }z=a_{1}x_{1}+a_{2}x_{2} + \cdots \ + a_{n}x_{n}$;
- Haja ao menos uma restri√ß√£o que n√£o seja expressa como uma equa√ß√£o ou inequa√ß√£o linear.

E existem v√°rios m√©todos para resolver um problema de PNL, sendo o mais abrangente as condi√ß√µes de [Karush-Kuhn-Tucker](https://pt.wikipedia.org/wiki/Condi%C3%A7%C3%B5es_de_Karush-Kuhn-Tucker) (ou simplesmente KKT) pelo fato de este m√©todo conseguir resolver PNL tanto com restri√ß√µes de igualdade quanto com os de desigualdade.

# Inst√¢ncia do Problema

> Uma ind√∫stria de papel est√° criando um novo modelo de folha com o intuito de atender o pedido de um cliente, que est√° requisitando um tipo n√£o padronizado de folha para o seu novo projeto de caderno. No entanto, o cliente n√£o informou quais seriam as dimens√µes desta folha de papel. Quando questionado, ele respondeu que as dimens√µes podem ser de quaisquer valores, desde que o per√≠metro da folha de papel seja igual a 200 cm, e de prefer√™ncia que possua a maior √°rea poss√≠vel. Com o intuito de atender a prefer√™ncia do cliente, deseja-se saber quais s√£o as dimens√µes do papel que maximizem a sua √°rea.

Exemplo retirado em: [M√âTODO DOS MULTIPLICADORES DE LAGRANGE üìö AULA 1](https://www.youtube.com/watch?v=iPyMGocdQyU)

---

O problema √© um cl√°ssico exemplo de PNL, pois a fun√ß√£o objetivo, que √© a √°rea da folha de papel, √© uma fun√ß√£o-n√£o linear por se tratar do produto entre duas vari√°veis de decis√£o, neste caso as dimens√µes $x$ e $y$ da folha de papel.

$$
\text{Max } √Årea(x,y)=xy \\
\begin{array}{rll}
  \text{Sujeito a: } & & \\
  & x + y = 100 & (1) \\
  & x, y > 0  & (2)
\end{array}
$$

Onde:

- As vari√°veis $x$ e $y$ respresentar√£o as dimens√µes da folha de papel dadas em cm;
- A restri√ß√£o $(1)$ √© a simplifica√ß√£o da restri√ß√£o do per√≠metro dado pelo cliente (Como o per√≠metro da folha dever ser de 200 cm, significa que $2x+2y=200 ‚Üí x+y=100$);
- A restri√ß√£o $(2)$ garante a positividade das vari√°veis $x$ e $y$.

# Resolvendo com Pyomo

```python
import pyomo.environ as pyo # Importando o Pyomo

# MODELO
modelo = pyo.ConcreteModel() # Criando o modelo concreto para o nosso problema

# VARI√ÅVEIS
modelo.x = pyo.Var(domain=pyo.PositiveReals) # Definindo a vari√°vel x
modelo.y = pyo.Var(domain=pyo.PositiveReals) # Definindo a vari√°vel y
x = modelo.x # "Renomeando" a vari√°vel x
y = modelo.y # "Renomeando" a vari√°vel y

# FUN√á√ÉO OBJETIVA
modelo.Area = pyo.Objective(expr= x*y, sense=pyo.maximize) # Definindo a fun√ß√£o objetivo
Area = modelo.Area # "Renomeando" a fun√ß√£o √Årea

# RESTRI√á√ÉO
modelo.Perimetro = pyo.Constraint(expr= x+y==100) # Definindo a restri√ß√£o simplificada do per√≠metro

# SOLUCIONANDO O MODELO
ipopt = pyo.SolverFactory('ipopt', executable='/content/ipopt') # Construindo o solver Ipopt
resultado = ipopt.solve(modelo) # Mandando o Ipopt resolver o modelo
```

E o resultado do nosso problema √© dado no bloco a seguir:

```python
# Se o status do resultado for 'ok', exibe as informa√ß√µes
if str(resultado.Solver.status) == 'ok':
  print(f"√Årea m√°xima = {value(Area):.2f} cm¬≤")
  print(f"x = {value(x):.2f} cm\ny = {value(y):.2f} cm")
else: # Caso contr√°rio...
  print("Nenhuma solu√ß√£o v√°lida foi encontrada.")
```

# Breve introdu√ß√£o ao SymPy

O SymPy (**Sym**bolic **Py**thon) √© uma biblioteca Python para matem√°tica simb√≥lica que visa se tornar um sistema de √°lgebra computacional (ou _Computer Algebric System_, ou ainda, CAS) completo, mantendo o c√≥digo o mais simples poss√≠vel para ser compreens√≠vel e facilmente extens√≠vel.

Um CAS, por sua vez, pode ser utilizado para c√°lculos com express√µes matem√°ticas complicadas, resolu√ß√£o de equa√ß√µes, simula√ß√µes, dentre outras aplica√ß√µes. H√° sistemas gratuitos como o SymPy e o Octave, e comerciais como Maple, MATLAB e Mathematica, cuja decis√£o de preferir um ou outro depende dos aspectos t√©cnicos e da an√°lise custo vs benef√≠cio do seu projeto. Para mais informa√ß√µes sobre a biblioteca, visite [o site do projeto](https://www.sympy.org/en/index.html).

Por se tratar de um CAS, o SymPy, ao contr√°rio de bibliotecas como Math cujos c√°lculos s√£o aproxima√ß√µes (o que pode levar a erros e acumulos de erros nas contas), representa n√∫meros e opera√ß√µes simbolicamente, o que permite que o resultado das contas seja obtida de forma **EXATA** (ou **analiticamente**, como √© formalmente dito).

## Usando SymPy para Verificar Convexidade de Fun√ß√µes

Uma coisa interessante de ver em PNL √© se uma dada fun√ß√£o √© estritamente convexa, convexa, c√¥ncava, estritamente c√¥ncava ou nem c√¥ncava nem convexa. E a forma mais geral de verificar a convexidade de uma fun√ß√£o √© a partir da seguinte regra:

Sendo $f(X = (x_1, x_2,..., x_n)):\mathbb{R}^{n} \rightarrow \mathbb{R}$, a sua matriz Hessiana $H[f(X)]$ ser√° dada por:

$$
H[f(X)] = D^{2}f(X) = \begin{bmatrix}
  \frac{\partial^{2} f(X)}{\partial x_1^{2}} &
  \frac{\partial^{2} f(X)}{\partial x_1 \partial x_2} &
  \dots &
  \frac{\partial^{2} f(X)}{\partial x_1\partial x_n} \\
  \frac{\partial^{2} f(X)}{\partial x_2\partial x_1} &
  \frac{\partial^{2} f(X)}{\partial x_2^{2}} &
  \dots &
  \frac{\partial^{2} f(X)}{\partial x_1\partial x_n} \\
  \vdots & \vdots & \ddots & \vdots \\
  \frac{\partial^{2} f(X)}{\partial x_n\partial x_1} &
  \frac{\partial^{2} f(X)}{\partial x_n\partial x_2} &
  \dots &
  \frac{\partial^{2} f(X)}{\partial x_n^{2}}
\end{bmatrix} 
$$

Por sua vez, os menores principais l√≠deres de $H[f(X)]$ ser√£o definidos por:

$$
\begin{array}{lcl}
H_1 & = & \begin{bmatrix}
  \frac{\partial^{2} f(X)}{\partial x_1^{2}}
\end{bmatrix} \\
H_2 & = & \begin{bmatrix}
  \frac{\partial^{2} f(X)}{\partial x_1^{2}} &
  \frac{\partial^{2} f(X)}{\partial x_1 \partial x_2} \\
  \frac{\partial^{2} f(X)}{\partial x_2\partial x_1} &
  \frac{\partial^{2} f(X)}{\partial x_2^{2}}
\end{bmatrix} \\
H_3 & = & \begin{bmatrix}
  \frac{\partial^{2} f(X)}{\partial x_1^{2}} &
  \frac{\partial^{2} f(X)}{\partial x_1 \partial x_2} &
  \frac{\partial^{2} f(X)}{\partial x_1 \partial x_3}\\
  \frac{\partial^{2} f(X)}{\partial x_2\partial x_1} &
  \frac{\partial^{2} f(X)}{\partial x_2^{2}} &
  \frac{\partial^{2} f(X)}{\partial x_2\partial x_3} \\
  \frac{\partial^{2} f(X)}{\partial x_3\partial x_1} &
  \frac{\partial^{2} f(X)}{\partial x_3\partial x_2} &
  \frac{\partial^{2} f(X)}{\partial x_3^{2}}
\end{bmatrix} \\
& \vdots & \\
H_{k \le n} & = & \begin{bmatrix}
  \frac{\partial^{2} f(X)}{\partial x_1^{2}} &
  \frac{\partial^{2} f(X)}{\partial x_1 \partial x_2} &
  \dots &
  \frac{\partial^{2} f(X)}{\partial x_1\partial x_k} \\
  \frac{\partial^{2} f(X)}{\partial x_2\partial x_1} &
  \frac{\partial^{2} f(X)}{\partial x_2^{2}} &
  \dots &
  \frac{\partial^{2} f(X)}{\partial x_1\partial x_k} \\
  \vdots & \vdots & \ddots & \vdots \\
  \frac{\partial^{2} f(X)}{\partial x_k\partial x_1} &
  \frac{\partial^{2} f(X)}{\partial x_k\partial x_2} &
  \dots &
  \frac{\partial^{2} f(X)}{\partial x_k^{2}}
\end{bmatrix}
\end{array}
$$

E as determinantes $|H_1|, |H_2|, \dots, |H_n|$ dos menores principais l√≠deres determinar√£o se a matriz $H[f(X)]$ √© positiva definida, positiva semidefinida, negativa definida, negativa semidefinida ou indefinida segundo os seguintes casos:

- $H[f(X)]$ √© **positiva definida** se, e somente se, as determinantes de todos os seus menores principais l√≠deres forem positivas.
- $H[f(X)]$ √© **positiva semidefinida** se, e somente se, as determinantes de todos os seus menores principais l√≠deres forem positivas ou nulas.
- $H[f(X)]$ √© **negativa definida** se, e somente se, as determinantes de todos os seus menores principais l√≠deres de ordem √≠mpar forem negativas e as de ordem par forem positivas.
- $H[f(X)]$ √© **negativa semidefinida** se, e somente se, as determinantes de todos os seus menores principais l√≠deres de ordem √≠mpar forem negativas ou nulas e as de ordem par forem positivas ou nulas.
- $H[f(X)]$ √© **indefinida** se n√£o cair em nenhum dos casos anteriores.

Finalmente, dada todas as defini√ß√µes at√© aqui, a convexidade de $f(X^{*})$, num dado ponto cr√≠tico $X^{*}=(x_1^*, x_2^*, \dots, x_n^*)$, ser√° dada conforme os casos a seguir:

- $f(X^{*})$ ser√° estritamente convexa se, e somente se, $H[f(X^{*})]$ for positiva definida.
- $f(X^{*})$ ser√° convexa se, e somente se, $H[f(X^{*})]$ for positiva semidefinida.
- $f(X^{*})$ ser√° estritamente c√¥ncava se, e somente se, $H[f(X^{*})]$ for negativa definida.
- $f(X^{*})$ ser√° c√¥ncava se, e somente se, $H[f(X^{*})]$ for negativa semidefinida.
- $f(X^{*})$ n√£o ser√° c√¥ncava nem convexa se, e somente se, $H[f(X^{*})]$ for indefinida.

A tabela a seguir resume todos os casos:

<center>

|Condi√ß√£o|Matriz $H[f(X^{*})]$|$f(X^{*})$|
|:-------|:-------------------|:---------|
|$|H_1|>0, |H_2|>0, |H_3|>0, \dots$|Positiva definida|Estritamente convexa|
|$|H_1| \ge 0, |H_2| \ge 0, |H_3|\ge 0, \dots$|Positiva semidefinida|Convexa|
|$|H_1|<0, |H_2|>0, |H_3|<0, \dots$|Negativa definida|Estritamente C√¥ncava|
|$|H_1| \le 0, |H_2| \ge 0, |H_3| \le 0, \dots$|Negativa semidefinida|C√¥ncava|
|Nenhuma das condi√ß√µes anteriores √© satisfeita|Indefinida|Nem c√¥ncava nem convexa|

</center>

E a partir dos conhecimentos apresentados at√© aqui, utilizando a biblioteca SymPy, vamos construir uma fun√ß√£o em Python que determinar√° a convexidade de uma express√£o _f_ a partir de sua matriz Hessiana e do c√°lculo da determinante dos seus menores principais l√≠deres, utilizando a fun√ß√£o `hessian()` e o m√©todo `.det()`, respectivamente.

```python
def convexidade(funcao, ponto: dict): # NOTA: Essa function s√≥ funciona se a determinante tiver APENAS CONSTANTES.

  from sympy import hessian, pprint, Function # Importando a fun√ß√£o para calcular a matriz Hessiana
  f = Function('f')
  f = funcao

  vars = [var for var in f.free_symbols] # Lista de vari√°veis de f.
  H = hessian(f, vars) # Matriz Hessiana
  print("Matriz Hessiana")
  pprint(H)

  Dets = [] # Array que armazenar√° as determinantes dos menores principais l√≠deres
  for i in range(1, len(vars)+1):
    Dets.append(H[:i,:i].det().subs(ponto))

  estritamente_convexa = [Dets[i] >  0 for i in range(len(Dets))]
  convexa              = [Dets[i] >= 0 for i in range(len(Dets))]
  estritamente_concava = [Dets[i] <  0 if i%2==0 else Dets[i] >  0 for i in range(len(Dets))]
  concava              = [Dets[i] <= 0 if i%2==0 else Dets[i] >= 0 for i in range(len(Dets))]

  if all(estritamente_convexa):
    return f"A fun√ß√£o √© estritamente convexa em {ponto}."
  elif all(convexa):
    return f"A fun√ß√£o √© convexa em {ponto}."
  elif all(estritamente_concava):
    return f"A fun√ß√£o √© estritamente c√¥ncava em {ponto}."
  elif all(concava):
    return f"A fun√ß√£o √© c√¥ncava em {ponto}."
  else:
    return f"A fun√ß√£o n√£o √© convexa nem c√¥ncava em {ponto}."
```

## Usando SymPy para resolver PNL com as condi√ß√µes de KKT

As condi√ß√µes de KKT representam o m√©todo mais abrangente de resolver problemas de PNL por poderem lidar com restri√ß√µes de igualdade e de desigualdade, cuja formula√ß√£o √© dada a seguir:

Seja $f(x = (x_1, x_2,\dots, x_n)):\mathbb{R}^{n} \to \mathbb{R}$ uma fun√ß√£o escalar de classe $C^1$ cujo problema de otimiza√ß√£o √© dado por:

$$
\text{Max } z=f(x)
$$

$$
\begin{array}{ll}
  \text{S.a.: } & \\
  & H = \left\{\begin{array}{l}
        h_1(x)=0 \\
        h_2(x)=0 \\
        \vdots \\
        h_m(x)=0
        \end{array}\right. \\
  & G = \left\{\begin{array}{l}
        g_1(x) \le 0 \\
        g_2(x) \le 0 \\
        \vdots \\
        g_k(x) \le 0
        \end{array}\right.
\end{array}
$$

Onde os conjuntos $H$ e $G$ s√£o restri√ß√µes de igualdade e desigualdade, respectivamente, definidas por fun√ß√µes escalares $h, g: \mathbb{R}^n \to \mathbb{R}$, taamb√©m de classe $C^1$, que cumprem as codi√ß√µes de regularidade em todos os pontos, isto √©, a jacobiana do conjunto de restri√ß√µes ativas $R=H \cup G$ √© linearmente indenpendente. Em termos mais formais, o posto de $J_R(x)$ √© igual a $m+k$.

$$
J_R(x)=\begin{bmatrix}
  \nabla h_1(x) \\
  \nabla h_2(x) \\
  \vdots \\
  \nabla h_m(x) \\
  \nabla g_1(x) \\
  \nabla g_2(x) \\
  \vdots \\
  \nabla g_k(x) \\
\end{bmatrix}
=\begin{bmatrix}
  \frac{\partial h_1(x)}{\partial x_1} &
  \frac{\partial h_1(x)}{\partial x_2} &
  \dots &
  \frac{\partial h_1(x)}{\partial x_n} \\
  \frac{\partial h_2(x)}{\partial x_1} &
  \frac{\partial h_2(x)}{\partial x_2} &
  \dots &
  \frac{\partial h_2(x)}{\partial x_n} \\
  \vdots & \vdots & \ddots & \vdots \\
  \frac{\partial h_m(x)}{\partial x_1} &
  \frac{\partial h_m(x)}{\partial x_2} &
  \dots &
  \frac{\partial h_m(x)}{\partial x_n} \\
  \frac{\partial g_1(x)}{\partial x_1} &
  \frac{\partial g_1(x)}{\partial x_2} &
  \dots &
  \frac{\partial g_1(x)}{\partial x_n} \\
  \frac{\partial g_2(x)}{\partial x_1} &
  \frac{\partial g_2(x)}{\partial x_2} &
  \dots &
  \frac{\partial g_2(x)}{\partial x_n} \\
  \vdots & \vdots & \ddots & \vdots \\
  \frac{\partial g_k(x)}{\partial x_1} &
  \frac{\partial g_k(x)}{\partial x_2} &
  \dots &
  \frac{\partial g_k(x)}{\partial x_n}
\end{bmatrix}
\ne \overrightarrow{0}, \forall  X \in R
$$

Ent√£o existem constantes $\lambda_1, \lambda_2, \dots, \lambda_m$ e $\mu_1, \mu_2, \dots, \mu_k$, chamados de multiplicadores de KKT, que satisfazem um conjunto de restri√ß√µes que permitem encontrar os pontos cr√≠ticos do Lagrangeano $L(x_1, \dots, x_n, \lambda_1, \dots, \lambda_m, \mu_1, \dots, \mu_k): \mathbb{R}^{n+m_+k} \to \mathbb{R}$, que s√£o os candidatos a maximizarem $f(x)$.

Tal Lagrangeano √© dado por:

$$
L = f(x) - \sum_{i=1}^{m} \lambda_{i} h_{i}(x)-\sum_{j=1}^{k} \mu_{j} g_{j}(x)
$$

Cujo conjunto de restri√ß√µes, conhecido por **condi√ß√µes de KKT**, as quais permitem encontrar os pontos candidatos a maximizarem f(x), √© dado por.

$$
\left\{\begin{array}{lll}
  \nabla f(x) = \sum_{i=1}^{m} \lambda_{i} \nabla h_i(x) + \sum_{j=1}^{k} \mu_{j} \nabla g_j(x) & & (1) \\
  h_i(x) = 0, & \forall i=1,\dots,m & (2) \\
  g_j(x) \le 0, & \forall j=1, \dots, k & (3) \\
  \mu_j \cdot g_j(x)=0, & \forall j=1,\dots,k & (4) \\
  \mu_j \ge 0, & \forall j=1, \dots, k & (5)
  \end{array}\right.
$$

Note-se que as restri√ß√µes $(1)$ e $(2)$ s√£o obtidas ao se calcular $\frac{\partial L(x, \lambda, \mu)}{\partial x}$ e $\frac{\partial L(x, \lambda, \mu)}{\partial \lambda}$, respectivamente. E que caso $k=0$, isto √©, n√£o houvesse restri√ß√µes de desigualdade, as condi√ß√µes KKT se transformam nas condi√ß√µes de Lagrange e os multiplicadores KKT s√£o chamados multiplicadores de Lagrange.

No intuito de resolver os problemas de otimiza√ß√£o n√£o linear, foi implementado a fun√ß√£o `KKT()`, utilizando a biblioteca SymPy j√° apresentada, em que s√£o passadas como argumentos para essa fun√ß√£o _strings_ para a fun√ß√£o $f(x)$ a ser otimizada e as restri√ß√µes de igualdade e desigualdade j√° no formato $h(x)=0$ e $g(x) \le 0$, ambas passadas como duas listas de _strings_. Caso se busque minimizar ao inv√©s de maximizar a fun√ß√£o, o par√¢metro booleano `max` deve ser definida como `False`. Ao final, a fun√ß√£o `KKT()` retorna uma lista cujo primeiro item corresponde ao valor √≥timo encontrado para $f(x)$ e os itens restantes s√£o os pontos $p$ encontrados que maximizam $f(x)$.

Al√©m disso, foi criado a fun√ß√£o `print_solucao_KKT()` para dar uma melhor visualiza√ß√£o da solu√ß√£o encontrada pela fun√ß√£o `KKT()` quando imprim√°-la na tela.

```python
def KKT(f: str, h: list = [], g: list = [], max=True) -> list:

  # NOTA: As express√µes contidas em h e g devem ser tais que h=0 e g<=0

  import sympy as sp # Importando a biblioteca SymPy
  m = len(h) # Quantidade de restri√ß√µes de igualdade
  k = len(g) # Quantidade de restri√ß√µes de desigualdade

  if max == True:
    f = sp.parse_expr(f) # Construindo a fun√ß√£o f pela string passada como par√¢metro
  else:
    f = sp.parse_expr(f'-({f})')
  
  vars = f.free_symbols # Obtendo o conjunto de vari√°veis
  pars = list(vars) # Lista de par√¢metros

  if m > 0:
    for i in range(m):
      exec(f'h{i+1} = sp.Symbol("h_{i+1}")')
      exec(f'h{i+1} = sp.parse_expr(h[{i}])') # Cria a restri√ß√£o de igualdade h_i
      exec(f'l{i+1} = sp.Symbol("\\lambda_{i+1}")') # Cria o par√¢metro lambda_i
      exec(f'pars.append(l{i+1})') # Adiciona lambda_i √† lista de par√¢metros

  if k > 0:
    for j in range(k):
      exec(f'g{j+1} = sp.Symbol("g_{j+1}")')
      exec(f'g{j+1} = sp.parse_expr(g[{j}])') # Cria a restri√ß√£o de desigualdade g_i
      exec(f'u{j+1} = sp.Symbol("\\mu_{j+1}", negative=False)') # Cria o par√¢metro mu_i
      exec(f'p{j+1} = sp.Symbol("p_{j+1}", negative=False)') # Cria a folga p_i
      exec(f'pars.append(u{j+1})') # Adiciona mu_i √† lista de par√¢metros
      exec(f'pars.append(p{j+1})') # Adiciona p_i √† lista de par√¢metros

  # Criando o Lagrangeano
  L = sp.Symbol('L')
  L = f

  if m > 0:
    for i in range(m):
      locais_dict = dict() # Dicion√°rio de vari√°veis locais a ser passado para exec()
      exec(f'L = L - l{i+1}*h{i+1}',
           {'L': L,
            f'l{i+1}': sp.Symbol(f'\\lambda_{i+1}'),
            f'h{i+1}': sp.parse_expr(h[i])
           },
           locais_dict
          )
      L = locais_dict['L'] # Adicionando a parcela lambda_j * h_j ao Lagrangeano

  if k > 0:
    for j in range(k):
      locais_dict = dict() # Dicion√°rio de vari√°veis locais a ser passado para exec()
      exec(f'L = L - u{j+1}*g{j+1}',
           {'L': L,
            f'u{j+1}': sp.Symbol(f'\\mu_{j+1}', negative=False),
            f'g{j+1}': sp.parse_expr(g[j])
           },
           locais_dict
          )
      L = locais_dict['L'] # Adicionando a parcela mu_i * g_i ao Lagrangeano

  # Criando as condi√ß√µes de KKT a partir do Lagrangeano
  R = [] # Conjunto de restri√ß√µes

  for var in vars:
    R.append(sp.diff(L, var)) # Adiciona a derivada do Lagrangeano em rela√ß√£o √† vari√°vel 'var'

  if m > 0:
    for i in range(m):
      exec(f'R.append(h{i+1})')
  
  if k > 0:
    for j in range(k):
      exec(f'R.append(u{j+1} * g{j+1})')
      exec(f'R.append(g{j+1} + p{j+1})')

  # Resolvendo as condi√ß√µes de KKT
  crit_lagr = sp.solve(R, pars, dict=True) # Pontos cr√≠ticos do Lagrangeano
  pontos_f = [{var: ponto.get(var) for var in ponto if var in vars} for ponto in crit_lagr] # pontos no dom√≠nio de f

  # Buscando o(s) ponto(s) que maximiza(m) f

  maximo = -float('inf') # infinito negativo
  solucao = list()

  for ponto in pontos_f:
    if f.subs(ponto) > maximo:
      maximo = f.subs(ponto)
  
  if max:
    solucao.append({'valor √≥timo': maximo})
  else:
    solucao.append({'valor √≥timo': -maximo})
  
  for ponto in pontos_f:
    if f.subs(ponto) == maximo:
      solucao.append(ponto)

  return solucao

def print_solucao_KKT(solucao: list, vars: list):

  from IPython.display import display, Latex
  from sympy import latex, Symbol

  valor_otimo = latex(solucao[0]['valor √≥timo'])

  display(Latex('\\text{Valor √≥timo: } f(p^*)=' + valor_otimo))
  print('- '*50)
  display(Latex('\\text{Ponto(s)}'))

  for i in range(1, len(solucao)):

    ponto = f"$p_{i}({vars[0]}"
    for j in range(1, len(vars)):
      ponto += f",{vars[j]}"

    ponto += f")=({latex(solucao[i][Symbol(vars[0])])}"
    for j in range(1, len(vars)):     
      ponto += f",{latex(solucao[i][Symbol(vars[j])])}"
    ponto += ")$"

    display(Latex(ponto))
```
