# PROBLEMA DE DESIGNAÇÃO
---
---
<p align=justify> &emsp; O Problema de Designação (<i>Assignment Problem</i>) nada mais é do que uma extensão (ou melhor dizendo, um caso mais particular) do Problema de Transporte, pois da mesma forma que o Problema de Transporte, geralmente se tem um conjunto de $m$ funcionários (ou grupos de funcionários) que devem ser designados em $n$ tarefas, onde cada tarefa $j$ deve ser realizada por um e somente um funcionário $i$ e cada trabalhador (ou grupo) $i$ deve ser designado a uma, e somente uma, tarefa $j$.
<br>
&emsp; Porém, na prática, não há necessidade de haver essa restrição em todos os Problemas de Designação, podendo haver possibilidade de uma tarefa $j$ ser designada a mais de um trabalhador (ou grupo), porém, na prática, é raro de haver necessidade de alocar um funcionário (ou grupo) $i$ realizando várias tarefas de uma vez, mas não é impossível. Por conta disso, será necessário informar qual o número $f_j$ de funcionários (ou grupos) que deverão ser designados para realizar a tarefa $j$ e o número de tarefas $t_i$ que devem ser designadas a cada funcionário (ou grupo) $i$. No caso particular em que cada funcionário (ou grupo) deva ser designado a realizar uma tarefa, e vice-versa, os valores de $f_j$ e $t_i$ serão todos iguais a 1.
<br>
&emsp; Porém para cada designação $(i,j)$, existe um custo associado $c_{ij}$ para designar a tarefa $j$ ao funcionário (ou grupo) $i$. Geralmente, esse custo é o tempo que o funcionário $i$ leva para finalizar a tarefa $j$, e o objetivo deste problema é minimizar o tempo total que os funcionários levam para realizar todas as atividades. Porém vale ressaltar que a designação de $n$ tarefas para $m$ funcionários não é o único tipo de instanciação para este problema, podendo utilizar o modelo do Problema de Designação para resolver, por exemplo, a minimização de custo de estoque em armazéns ("Quais materiais devem estar estocados em quais armazéns?") ou de despacho de veículos de tranporte das unidades para as distribuidores ("Quais caminhões devemos enviar para quais distribuidoras?").
<br>
&emsp; Fazendo um paralelo com o Problema de Trasnporte, os funcionários $i$ e as tarefas $j$ corresponderiam aos fornecedores $i$ e aos clientes $j$, onde o número máximo $s_i$ de unidades é o equivalente ao número $t_i$ de tarefas que cada funcionário está disposto a realizar, enquanto o número $d_j$ de unidades demandadas por cada cliente $j$ corresponde à quantidade $f_j$ de funcionários que cada tarefa $j$ pode designar. Por fim, o custo $c_{ij}$ por cada unidade produzida na fábrica $i$ e transportada ao cliente $j$ corresponde ao custo $c_{ij}$ por cada designação da tarefa $j$ ao funcionário $i$.
<br>
&emsp; Como pode ser visto a seguir, o modelo de programação linear para o Problema de Designação é bastante semelhante ao modelo de PL do Problema de Transporte, com a diferença de que as variáveis de decisão são binárias, e o objetivo é minimizar os custos das possíveis designações.
<br>
<br>
$$
\text{Min } z = \sum_{i=1}^{m} \sum_{j=1}^{n} c_{ij}x_{ij}
$$
<br>
$$
\begin{array}{rlll}
  \text{Sujeito a: } & & \\
  & \sum_{j=1}^{n} x_{ij} = t_{i} & , \forall \ i=1,...,m & (1) \\
  & \sum_{i=1}^{m} x_{ij} = f_{j} & , \forall \ j=1,...,n & (2) \\
  & x \in \{0,1\} & & (3)
\end{array}
$$
<br>
&emsp; Onde:
<ul align=justify>
<li>$x_{ij}$ é a decisão de designar o funcionário (ou grupo) $i$ para realizar a tarefa $j$;</li>
<li> $c_{ij}$ é o custo de designar o funcionário (ou grupo) $i$ à tarefa $j$;</li>
<li> $t_i$ corresponde ao número de tarefas que o funcinário $i$ deve relizar;</li>
<li> $f_j$ corresponde ao número de funcionários que a tarefa $j$ deverá necessitar para ser realizada;</li>
<li> O conjunto de restrições $(1)$ garante que todos os $n$ funcionários sejam designados a fazer todas as tarefas que puderem realizar, evitando que qualquer funcionário tenha capacidade ociosa;</li>
<li> O conjunto $(2)$ garante que todas as $m$ tarefas sejam designadas aos $n$ funcionários, evitando haver alguma atividade sem designação;</li>
<li>A restrição $(3)$ garante a binaridade dos $x_{ij}$.</li>
</ul>
</p>

# Resolução
---

<p align=justify> &emsp; Vamos considerar que os custos de transporte das máquinas até as fábricas estejam organizados em uma matriz e o número de designações de cada uma das máquinas e de cada uma das fábricas estejam armazenados em duas listas distintas, como mostrado a seguir:</p>

```python
maquinas = ['1', '2', '3', '4'] # Nome das máquinas
fabricas = ['1', '2', '3', '4'] # Nome das fábricas

f_j = [1, 1, 1, 1] # número de designações que cada máquina deve fazer.
t_i = [1, 1, 1, 1] # número de designações que cada fábrica deve fazer.

# Custos c_ij de designar a máquina i para a fábrica j:

#        F1  F2  F3  F4
c_ij = [[10, 12, 15, 16], # M1
        [14, 12, 13, 18], # M2
        [10, 16, 19, 15], # M3
        [14, 12, 13, 15]] # M4
```

<p align=justify> &emsp; Para a resolução do problema instanciado com o Pyomo, é possível criar uma função que importe a biblioteca do Pyomo, e resolva o problema de transporte usando como parâmetros os dados fornecidos nos argumentos desta função, cujo código é mostrado como se segue:</p>

```python
def problema_designação(ofertantes, ofertas, demandantes, demandas, custos):

  import pyomo.environ as pyo # Importando o Pyomo

  modelo = pyo.ConcreteModel()           # Criando uma instância do modelo
  modelo.I = pyo.RangeSet(len(ofertantes))  # Índice para as máquinas
  modelo.J = pyo.RangeSet(len(demandantes)) # Índice para as fábricas
  I = modelo.I
  J = modelo.J
  
  # VARIÁVEIS DE DECISÃO
  # xij representa a decisão de designar a máquina i para a fábrica j
  modelo.x = pyo.Var(I, J, within= pyo.Binary)
  x = modelo.x
  
  # PASSANDO OS PARÂMETROS DO PROBLEMA

  # Custos de designar a máquina i para a fábrica j
  modelo.c = pyo.Param(I, J, initialize = lambda modelo, i, j: custos[i-1][j-1])
  c = modelo.c

  # Número de designações para cada máquina
  modelo.s = pyo.Param (I, initialize = lambda modelo, i: ofertas[i-1])
  s = modelo.s

  # Número de designações para cada fábrica
  modelo.d = pyo.Param (J, initialize = lambda modelo, j: demandas[j-1])
  d = modelo.d

  # FUNÇÃO OBJETIVO
  modelo.obj = pyo.Objective(
      rule= lambda modelo: sum(x[i, j] * c[i, j] for i in I for j in J),
      sense= pyo.minimize
      )
  obj = modelo.obj

  # RESTRIÇÕES

  # Conjunto 1 de Restrições (R1):
  # O total de designações de cada máquina deve ser igual ao seu número permissível de designações.
  modelo.R1 = pyo.Constraint(
      I,
      rule= lambda modelo, i: sum(x[i, j] for j in J) == s[i]
      )
  
  # R2: O total de máquinas designadas a cada fábrica deve ser igual ao número permitido de designações a essa fábrica
  modelo.R2 = pyo.Constraint(
      J,
      rule= lambda modelo, j: sum(x[i, j] for i in I) == d[j]
      )
  
  # RESOLUÇÃO DO MODELO
  gurobi = pyo.SolverFactory('gurobi') # Construindo o solver gurobi
  resultado = gurobi.solve(modelo) # Armazenando o resultado
  
  if str(resultado.Solver.status) == 'ok':

    # Imprimindo a função objetiva
    registro = f"CUSTO MÍNIMO\nz = {pyo.value(obj)}\n"

    # Imprimindo as variáveis de decisão selecionadas
    chaves = list(modelo.x.keys())
    registro += '\nVARIÁVEIS SELECIONADAS'
    for c in chaves:
      valor = pyo.value(x[c])
      if valor != 0:
        registro += f'\nx{ofertantes[c[0]-1]}{demandantes[c[1]-1]}'

  else:

    registro = 'Nenhuma solução válida foi encontrada.'

  return registro

  """Perceba que este código é praticamente igual ao do Problema de Transporte,
  com a única diferença de que as restrições não são mais inequações, mas sim equações."""
```

<p align=justify> &emsp; Cujo resultado encontrado será como dado a seguir:</p>

```python
registro = problema_designação(maquinas, t_i, fabricas, f_j, c_ij)
print(registro)
```

# RESOLVENDO COM NETWORKX
---
---

<p align=justify>
&emsp; A biblioteca NetworkX conta com a função <a href='https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.bipartite.matching.minimum_weight_full_matching.html'><code>minimum_weight_full_matching()</code></a>, que se encontra dentro do módulo <a href='https://networkx.org/documentation/stable/reference/algorithms/bipartite.html'><code>bipartite</code></a>. Esta função retorna as arestas de um grafo bipartido que possuem a menor soma de pesos que associam os nós de uma partição aos nós da outra partição, o que bate exatamente com as características do Problema de Designação lidado.
<br>
&emsp; Mas antes, devemos realizar a construção do grafo antes de apliar esta função.
</p>

```python
import networkx as nx

# Instanciando um grafo direcionado
G = nx.DiGraph()

# Criando os nós a partir de uma lista
G.add_nodes_from(["M1", "M2", "M3", "M4", "F1", "F2", "F3", "F4"])

# Criando as arestas a partir de uma lista
# Foi definido um atributo 'custo' para cada aresta para ser ultizado em min_cost
G.add_edges_from([
    ("M1", "F1", {'custo': 10}),
    ("M1", "F2", {'custo': 12}),
    ("M1", "F3", {'custo': 15}),
    ("M1", "F4", {'custo': 16}),
    ("M2", "F1", {'custo': 14}),
    ("M2", "F2", {'custo': 12}),
    ("M2", "F3", {'custo': 13}),
    ("M2", "F4", {'custo': 18}),
    ("M3", "F1", {'custo': 10}),
    ("M3", "F2", {'custo': 16}),
    ("M3", "F3", {'custo': 19}),
    ("M3", "F4", {'custo': 15}),
    ("M4", "F1", {'custo': 14}),
    ("M4", "F2", {'custo': 12}),
    ("M4", "F3", {'custo': 13}),
    ("M4", "F4", {'custo': 15})
    ])

# Constroi o dicionario com as designações
desig_dict = nx.bipartite.minimum_weight_full_matching(
    G,
    top_nodes=["M1", "M2", "M3", "M4"],
    weight='custo'
)
print(f'Designações: {desig_dict}')
```

<p align=justify>
&emsp; Uma outra forma de resolver este problema seria também usando, mais uma vez, as funções <a href='https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.flow.min_cost_flow.html'><code>min_cost_flow()</code></a> e <a href='https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.flow.min_cost_flow_cost.html'><code>min_cost_flow_cost()</code></a>, como foi apresentado nas duas aulas anteriores. Para isso, vamos usar a função <a href='https://networkx.org/documentation/stable/reference/generated/networkx.classes.function.set_node_attributes.html'><code>set_nodes_attributes()</code></a> e atribuir para cada máquina uma demanda $-1$, e para cada fábrica, uma demanda $+1$.
</p>

```python
# Criando o atributo 'demanda' para cada um dos nós de G
nx.set_node_attributes(
    G,
    {"M1": -1, "M2": -1, "M3": -1, "M4": -1,
     "F1":  1, "F2":  1, "F3":  1, "F4":  1},
    name='demanda'
    )

# Calcula as designações de cada máquina a cada fábrica
designacoes = nx.min_cost_flow(
    G,
    demand='demanda',
    weight='custo'
    )
print(f'Designações: {designacoes}')

# Calcula o valor do custo mínimo
custo_min = nx.min_cost_flow_cost(
    G,
    demand='demanda',
    weight='custo'
    )
print(f'Custo Mínimo: {custo_min}')
```

## Desenhando o grafo
---

<p align=justify>
&emsp; Por motivos óbvios, foi utilizado a função <a href='https://networkx.org/documentation/stable/reference/generated/networkx.drawing.layout.bipartite_layout.html'><code>bipartite_layout()</code></a> para desenhar o grafo bipartido do nosso Problema de Designação.
</p>

```python
import matplotlib.pyplot as plt

# Definindo a disposição dos nós do grafo G
posicao = nx.bipartite_layout(
    G,
    nodes=["M1", "M2", "M3", "M4"],
    align='vertical'
    )

# Desenha o grafo G original
nx.draw_networkx(
    G,                       # Grafo original # Título do Grafo
    ax= plt.axes([0,0,1,1]), # Define o eixo de exibição
    pos= posicao,            # Define o layout dos nós         
    node_color= 'darkblue',  # Define a cor dos nós como azul escuro
    node_size= 500,          # Define o tamanho dos nós
    node_shape= 's',         # Define o formato dos nós como quadrado
    font_weight= 'bold',     # Coloca negrito no texto dos nomes dos nós
    font_color= 'white'      # Define a cor do texto dos nomes como branco
    )
plt.title('Grafo original')

# Desenhando as Designações
nx.draw_networkx(
    G,                       # Grafo original
    edgelist= [
        (no1, no2) for no1 in designacoes for no2 in designacoes[no1] if designacoes[no1][no2] == 1
        ],                   # Exibe apenas as designações encontradas
    ax= plt.axes([1,0,1,1]), # Define o eixo de exibição
    pos= posicao,            # Define o layout dos nós         
    with_labels= True,       # Os nós terão rótulos para exibir os seus nomes
    node_color= 'red',       # Define a cor dos nós como vermelho
    edge_color= 'darkred',     # Define a cor das arestas como vermelho escuro
    node_size= 500,          # Define o tamanho dos nós
    node_shape= 's',         # Define o formato dos nós como círculo
    font_weight= 'bold',     # Coloca negrito no texto dos nomes dos nós
    font_color= 'white'      # Define a cor do texto dos nomes como branco
    )
plt.title('Designações')
plt.legend(
    labels=[f'Custo Mínimo: R$ {custo_min:.2f}'],
    loc='lower center',
    framealpha= 0
    )
```

<p align=justify>
&emsp; Perceba que, nesta segunda forma de resolver o Problema de Designação, foi encontrado uma outra solução que permite alcançar o mesmo resultado ótimo de 50.
